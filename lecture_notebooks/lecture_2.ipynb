#### Coercivity of the variational problem

In order to apply the theory we have developed in the previous term, we need to show that the bilinear form $a(\cdot, \cdot)$ is coercive (it is clear that it is bounded on $H^1$).

From the approximation theory we have developed, we know that (with some minor conditions on the domain $\Omega$) there exists a constant $C$ depending on $\Omega$ such that
$$
   \inf_{r \in \mathbb{R}} \| v - r \|_{L^2(\Omega)} \leq C_\Omega | v |_{H^1(\Omega)}
$$
for all $v \in H^1(\Omega)$.

Also note that
$$
   \| v - \overline{v} \|_{L^2(\Omega)} = \inf_{r \in \mathbb{R}} \| v - r \|_{L^2(\Omega)}
$$
Why?

Consider first the pure Neumann problem.
Then, $v \in V$ implies $\overline{v} = 0$, and so
$$
   \| v \|_{L^2(\Omega)} \leq C_\Omega |v |_{H^1(\Omega)},
$$
and the coercivity result follows.

Now, suppose we have non-empty region $\Gamma_D$ with Dirichlet conditions.
We first will prove that
$$
   \| v \|_{L^2(\Omega)} \leq C \left( \int_{\Gamma_D} v \, ds + |u|_{H^1(\Omega)} \right)
$$

Let $v \in H^1(\Omega)$ be arbitrary.
Then,
$$
   \begin{aligned}
      \| v \|_{L^2(\Omega)}
         &\leq \| v - \overline{v} \|_{L^2(\Omega)} + \| \overline{v} \|_{L^2(\Omega)} \\
         &\leq C_\Omega | v |_{H^1(\Omega)} + \frac{|\Omega|^{1/2}}{|\Gamma_D|} \left| \int_{\Gamma_D} \overline{v} \, ds \right| \\
         &\leq C_\Omega | v |_{H^1(\Omega)} + \frac{|\Omega|^{1/2}}{|\Gamma_D|} \left(\left| \int_{\Gamma_D} v \, ds \right| + \left| \int_{\Gamma_D} (v - \overline{v}) \, ds \right| \right)
   \end{aligned}
$$

Note that
$$
   \begin{aligned}
      \left| \int_{\Gamma_D} (v - \overline{v}) \, ds \right|
         &\leq |\Gamma_D|^{1/2} \| v - \overline{v} \|_{L^2(\partial\Omega)} \\
         &\leq |\Gamma_D|^{1/2} C_\Omega \left( \| v - \overline{v} \|_{H^1(\Omega)} \right) \\
         &\leq |\Gamma_D|^{1/2} C_\Omega | v - \overline{v} |_{H^1(\Omega)}
   \end{aligned}
$$
by the trace theorem (and using again that $v - \overline{v}$ is zero-mean).

Coercivity follows immediately using that $v \in V$:

> **_PROPOSITION:_** For all $v \in V$, there exists a constant $C$ depending on $\Omega$ (and $\Gamma_D$ in the non-pure Neumann case) such that
> $$ a(v, v) \geq C \| v \|_{H^1(\Omega)} $$

In the pure Dirichlet case, this result is known as the **Poincaré inequality**,

> **_PROPOSITION (Poincaré Inequality):_** There exists a constant $C$ such that
> $$ \| u \|_{W^1_p(\Omega)} \leq C | v \|_{W^1_p(\Omega)} $$

#### Discrete approximations, error estimates

We have shown that $a(\cdot, \cdot)$ is bounded and coercive, i.e.
$$
\begin{aligned}
   a(v, w) &\leq C \| v \|_{H^1(\Omega)} \| w \|_{H^1(\Omega)} \\
   a(v, v) &\geq c \| v \|_{H^1(\Omega)}
\end{aligned}
$$

Given $f \in L^2(\Omega)$, let $u$ satisfy the variational problem in $H^1(\Omega)$, and let $u_h$ satisfy the variational problem restricted to $V_h$, a finite-dimensional subspace of $H^1$ (our discrete finite element space).

Then, Galerkin orthgonality tells us that
$$
   \| u - u_h \|_E = \inf_{v \in V_h} \| u - v \|_E,
$$
where
$$
   \| v \|_E = a(v, v) = \| \nabla v \|_{L^2(\Omega)} = | v |_{H^1(\Omega)}
$$
is the energy norm.

By boundedness and coercivity, we have
$$
\begin{aligned}
   \| u - u_h \|_{H^1(\Omega)}
      &\leq \frac{1}{\sqrt{c}} a(u - u_h, u - u_h) \\
      &= \frac{1}{\sqrt{c}} \inf_{v \in V_h} a(u - v, u - v) \\
      &\leq \frac{\sqrt{C}}{\sqrt{c}} \inf_{v \in V_h} \| u - v \|_{H^1(\Omega)}, \\
\end{aligned}
$$

We proved (the Bramble-Hilbert lemma) that for finite element spaces with polynomial degree $p$, we have the estimate
$$
   \| u - \mathcal{I}_h u \|_{H^1(\Omega)} \leq C h^p | u |_{H^{p+1}(\Omega)}.
$$

Combining this result with the above, we obtain the error estimate
$$
   \| u - u_h \|_{H^1(\Omega)} \leq C h^p | u |_{H^{p+1}(\Omega)}.
$$

#### $L^2$ error estimates

Let $e := u - u_h$ denote the error.

Consider the **dual problem**

$$
\left\{
\begin{aligned}
    -\Delta w &= e \qquad\text{in $\Omega$} \\
    w &= 0 \qquad\text{on $\Gamma_D \subseteq \partial\Omega$} \\
    \frac{\partial w}{\partial \boldsymbol n} &= 0 \qquad\text{on $\Gamma_N = \partial\Omega \setminus \Gamma_D$}
\end{aligned}
\right.
$$
which corresponds to the variational problem of finding $w \in V$ such that, for all $v \in V$,
$$
   a(w, v) = (e, v).
$$

Then,
$$
\begin{aligned}
    \| u - u_h \|_{L^2(\Omega)}^2
        &= (u - u_h, u - u_h) \\
        &= a(w, u - u_h) \quad\text{since $w$ solves the variational problem} \\
        &= a(w - \mathcal{I}_h w, u - u_h) \quad\text{since $u - u_h$ is orthognal to $V_h$} \\
        &\leq C \| w - \mathcal{I}_h w \|_{H^1(\Omega)} \| u - u_h \|_{H^1(\Omega)} \quad\text{by boundedness} \\
        &\leq C h \| u - u_h \|_{H^1(\Omega)} | w |_{H^2(\Omega)}
\end{aligned}
$$

Now, we make the assumption (that can be proven, see section 5.5 of the textbook) that
$$
    | w |_{H^2(\Omega)} \leq C \| e \|_{L^2(\Omega)}.
$$
Then, combining these results, we have proven that
$$
\begin{aligned}
    \| u - u_h \|_{L^2(\Omega)}
        &\leq C h \| u - u_h \|_{H^1(\Omega)} \\
        &\leq C h^{p+1} \| u - u_h \|_{H^{p+1}(\Omega)}
\end{aligned}
$$
